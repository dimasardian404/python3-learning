```python
fhand = open("mbox.txt")
count = 0
for line in fhand:
    if line.startswith('From:'):
        print(line)
        count += 1
print(count)
```

```python
fhand = open("mbox.txt")
count = 0
for line in fhand:
    line = line.rstrip()
    if line.startswith('From:'):
        print(line)
        count += 1
print(count)
```

```python
fhand = open("mbox.txt")
for line in fhand:
    line = line.rstrip()
    if not line.startswith('From:'):
        continue
    print(line)
```

```python
fhand = open("mbox.txt")
count = 0
for line in fhand:
    line = line.rstrip()
    if not 'umich.edu' in line:
        continue
    print(line)
    count += 1
print(count)
```
