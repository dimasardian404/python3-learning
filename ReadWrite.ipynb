```python
fhandler = open("mbox.txt")
lines = fhandler.readlines()

count = 0
for line in lines:
    line = line.rstrip()
    if line.startswith('From:'):
        xfile = open("EmailFrom.txt",'a') #w = write dan a = append
        # xfile.write(line)
        xfile.write(line + '\n')
        count += 1
        xfile.close()
```
