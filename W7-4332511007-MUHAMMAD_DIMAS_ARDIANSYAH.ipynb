2. Tambahkan kode pada fungsi bernama build_sentence (info) yang menerima satu argumen yang berisi string dan mengembalikan kalimat yang dimulai dengan string yang diberikan dan diakhiri dengan string "adalah keuntungan menggunakan fungsi!"
3. Jalankan program dan perhatikan Semua funngsi bekerja bersama!

```python
# Tugas Fungis 0

# Modify this function to return a list of strings as defined above
def list_benefits():
    return ["Kode yang lebih teratur", "Kode yang lebih mudah dibaca", 
            "Penggunaan kembali kode yang lebih mudah", "Memungkinkan pemrogram untuk berbagi dan menghubungkan kode bersama"] 

# Ubah fungsi ini untuk menggabungkan ke setiap 'benefit' - " adalah keuntungan menggunakan fungsi!"
def build_sentence(build_sent):
    return build_sent + ' adalah keuntungan menggunakan fungsi!'
    

def name_the_benefits_of_functions():
    list_of_benefits = list_benefits()
    for benefit in list_of_benefits:
        print(build_sentence(benefit))

name_the_benefits_of_functions()
```
