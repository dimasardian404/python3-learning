```python
# Fungsi Fruitfull
# Fungsi dengan balikan nilai (return value)
# Fungsi dengan 1 argumen

def persegi(s):
    Luas = s * s
    return Luas
    # print(Luas)

# print(persegi(int(input())))
nilaiLuas = persegi(10)
print(nilaiLuas)

# persegi(10)
```

```python
# Fungsi multi argumen

def LuasPersegiPanjang(p,l):
    Luas = p * l
    return Luas
    # return "Luasnya adalah " + str(Luas)

inp_p = int(input('Input nilai Panjang')) 
inp_l = int(input('Input nilai Lebar'))

hasil = LuasPersegiPanjang(inp_p,inp_l)
print('Luas dari persegi panjang adalah ' + str(hasil))

# print(x)
# x = hasil * 3
```
